**2) заполним пропуски в `locality_name`**

*В названиях населённого пункта нужно избавиться от неявных дубликатов, а также небольшое количество пропусков в последствии можно отбросить так как они существенно не повлияют на исследование. В дополнение, мы отсортируем от лишних значений столбец `locality_name`, так как в нем есть неявные дубликаты.*

```python
#количество пропусков в названии населенного пункта
data['locality_name'].isna().sum()
```

```python
data = data.query('locality_name.isna() == False')
```

```python
#убираем все ненужные названия и неправильную запись населенного пункта
not_correct = ['посёлок ', 'поселок ', 'деревня ', 'городской ', 'при ', 'железнодорожной ' ,'станции ', 'городского ',
         'село ', 'коттеджный ', 'садоводческое ', 'садовое ', 'товарищество ', 'садовое ', 'станции ', 'типа ', 'некоммерческое ']
correct=''
data['locality_name'] = data['locality_name'].replace(not_correct, correct, regex=True)
```

```python
#выведем на экран все названия
with pd.option_context("display.max_rows", 1000): 
       display(data['locality_name'].value_counts().sort_values(ascending=True))
```

*В названиях населённого пункта нужно избавиться от неявных дубликатов, а также небольшое количество пропусков в последствии можно отбросить так как они существенно не повлияют на исследование. В дополнение, мы отсортируем от лишних значений столбец `locality_name`, так как в нем есть неявные дубликаты.*

```python
#количество пропусков в названии населенного пункта
data['locality_name'].isna().sum()
```

**3) Заполним пропуски в столбце `ceiling height`**

```python
#Проверим наличие пропусков в data в столбце ceiling_height
data['ceiling_height'].isna().sum()
```

```python
#уникальные значения в ceiling_height
data['ceiling_height'].unique()
```

Данные в высоте потолков имеют большое количество пропусков - 9195 пропуски появились, возможно, из-за халатности людей не внесших значения в высоту потолков, а также аномальные значения в виде 27,5 метров они возможно появились также из-за неправильного ввода значений. Большинство людей сдают все таки квартиру в многоэтажках, а не одноэтажные дома, в соответствии с нормативом, в самых теплых климатических зонах потолок должен располагаться на высоте не менее 2,7 метра, а для остальных районов минимальная высота потолка составляет 2,5 метра. Санкт-Петербург и прилегающие территории не относятся к теплым климатических зонам, так что мы заменим пропуски медианным значением высоты потолков.

```python
#заменим аномальные значения 
for i in data['ceiling_height']:
    data.loc[data['ceiling_height'] >= 10, 'ceiling_height'] = i/10
```

```python
#Заменим пропуски на медианное значение в ceiling_height по населенному пункту
for i in data['locality_name'].unique():
    median = data.loc[data['locality_name'] == i, 'ceiling_height'].median()
    data.loc[(data['ceiling_height'].isna()) & (data['locality_name'] == i),'ceiling_height'] = median
```

```python
#Заменим оставшиеся пропуски на медианное значение в ceiling_height
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

```python
#Проверим наличие пропусков в data в столбце ceiling_height
data['ceiling_height'].isna().sum()
```

## Расширение таблицы

### Добавление столбцов 

**1)Добавим стобец - цена одного квадратного метра**

```python
#стоимость одного квадратного метра
good_data['cost_one_metr'] = good_data['last_price'] / good_data['total_area']
```

```python
good_data.head()
```

**2)Добавим стобцы - день недели месяц и год публикации**

```python
good_data.info()
```

```python
good_data['first_day_exposition'] = good_data['first_day_exposition'].astype('datetime64[M]')
```

```python
#добавим столбец день недели публикации
good_data['day_week'] = good_data['first_day_exposition'].dt.weekday
```

```python
good_data['month'] = good_data['first_day_exposition'].dt.month
```

```python
#добавим столбец год публикации
good_data['year'] = good_data['first_day_exposition'].dt.year
```

**3)Добавим столбец тип этажа**

```python
def floor_type(row):
        if row['floor'] == 1:
            return 'Первый'
        if row['floor'] == row['floors_total']:
            return 'Последний'
        return 'Другой'
good_data['floor_type'] = good_data.apply(floor_type, axis = 1)
```

**4)Добавим стообец расстояние до центра города**

```python
good_data.head()
```

```python
#добавляем столбец округленный до целых
good_data['distance_to_centre'] = round(good_data['cityCenters_nearest']/1000)
```

**Заполнение пропусков в `days_exposition`.** 

```python
good_data['year'].unique()
```

```python
#заполним пропуски в days_exposition
for i in good_data['year'].unique():
    good_data.loc[(good_data['year'] == i) & (good_data['days_exposition'].isna()), 'days_exposition'] = \
             good_data.loc[(good_data['year'] ==i), 'days_exposition'].median()
```
