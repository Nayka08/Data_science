Технологический процесс

Rougher feed — исходное сырье

Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant

Xanthate — ксантогенат (промотер, или активатор флотации);

Sulphate — сульфат (на данном производстве сульфид натрия);

Depressant — депрессант (силикат натрия).

Rougher process (англ. «грубый процесс») — флотация

Rougher tails — отвальные хвосты

Float banks — флотационная установка

Cleaner process — очистка

Rougher Au — черновой концентрат золота

Final Au — финальный концентрат золота

Параметры этапов:

air amount — объём воздуха

fluid levels — уровень жидкости

feed size — размер гранул сырья

feed rate — скорость подачи

## Построение моделей

Для решения задачи найдем — метрику качества sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).

```python
columns = ["final.output.recovery","rougher.output.recovery"]

test_df = test_df.join(full_df[columns])

for i in train_df.columns:
    if i not in test_df.columns:
        train_df = train_df.drop(i, axis = 1)
```

```python
for i in test_df:
    test_df[i] = test_df[i].interpolate()
```

```python
def smape(A, F):
    return 100/len(A) * np.sum(2 * np.abs(F - A) / (np.abs(A) + np.abs(F)))
```

```python
smape_score = make_scorer(smape, greater_is_better=True)
```

```python
features_train = train_df.drop(["rougher.output.recovery", "final.output.recovery"], axis=1)
target_train_rougher = train_df["rougher.output.recovery"]
target_train_final = train_df["final.output.recovery"]

features_test = test_df.drop(["rougher.output.recovery", "final.output.recovery"], axis=1)

target_test_rougher = test_df["rougher.output.recovery"]
target_test_final = test_df["final.output.recovery"]
```

Нужно спрогнозировать сразу две величины:

- эффективность обогащения чернового концентрата rougher.output.recovery;

- эффективность обогащения финального концентрата final.output.recovery.

Итоговая метрика складывается из двух величин:

### Модель - случайный лес

```python
pd.options.mode.chained_assignment = None

scaler = StandardScaler()
scaler.fit(features_train) 

features_train_rougher = scaler.transform(features_train) 

scaler.fit(features_test)

features_test = scaler.transform(features_test) 
```

```python
model_rf = RandomForestRegressor(random_state=12345)
    

param_grid_rf = {
    'n_estimators': np.arange(11, 101, 5),
    'max_depth': np.arange(2, 20),
    "min_samples_split" : [2,4,8],
}
```

```python
gs_rf = RandomizedSearchCV(
    estimator=model_rf,
    param_distributions=param_grid_rf, 
    scoring=smape_score, 
    n_jobs=-1, 
    random_state=12345,
    cv=5,
    error_score='raise'
)

gs_rf.fit(features_train, target_train_rougher)

best_estimator_rf = gs_rf.best_estimator_
best_score_rf = gs_rf.best_score_
best_params_rf = gs_rf.best_params_
print('Лучшие значения Случайного леса: {}'.format(best_score_rf))
      
print('Лучшие параметры Случайного леса: {}'.format(best_params_rf))
```

```python
gs_rf.fit(features_train, target_train_final)

best_estimator_rf_final = gs_rf.best_estimator_
best_score_rf_final = gs_rf.best_score_
best_params_rf_final = gs_rf.best_params_
print('Лучшие значения Случайного леса: {}'.format(best_score_rf_final))
      
print('Лучшие параметры Случайного леса: {}'.format(best_params_rf_final))
```

```python
final_smape_rf = 0.25*best_score_rf + 0.75*best_score_rf_final
final_smape_rf 
```

### Модель линейная регрессия

```python
model_lr = LinearRegression()
scores_rough = cross_val_score(model_lr, features_train, target_train_rougher, 
                                 scoring=smape_score, cv=5)
result_rough = sum(scores_rough) / len(scores_rough) 

scores_rough = cross_val_score(model_lr, features_train, target_train_final, 
                                 scoring=smape_score, cv=5)
result_final = sum(scores_rough) / len(scores_rough) 

final_smape_lr = 0.25*result_rough + 0.75*result_final
final_smape_lr
```

### Проверка модели

```python
model =  RandomForestRegressor(random_state = 12345, max_depth = 19, n_estimators = 21, n_jobs=-1) 
model.fit(features_train, target_train_rougher)
z = model.predict(features_test)
score_1 = smape(target_test_rougher, z)
score_1
```

```python
model =  RandomForestRegressor(random_state = 12345, max_depth = 19, n_estimators = 21, n_jobs=-1) 
model.fit(features_train, target_train_final)
score_2 = smape(target_test_final, z)
score_2
```

```python
final_score = 0.25*score_1 + 0.75*score_2
final_score
```

```python
dummy_regr = DummyRegressor(strategy="mean")
dummy_regr.fit(features_train, target_train_rougher)
d_p_rough = dummy_regr.predict(features_test)

dummy_score_1 = smape(target_test_rougher, d_p_rough)
dummy_score_1
```

```python
dummy_regr = DummyRegressor(strategy="mean")
dummy_regr.fit(features_train, target_train_final)
d_p_final = dummy_regr.predict(features_test)

dummy_score_2 = smape(target_test_final, d_p_final)
dummy_score_2
```

```python
final_dummy_score = 0.25*dummy_score_1 + 0.75*dummy_score_2
final_dummy_score
```
