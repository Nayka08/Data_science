## **Подготовка данных**

### Проверка на пропуски и приведение данных в нижние регистры

Приведем назвния столбцов к нижним регистрам

```python
#приведем названия столбцов к нижним регистрам
data.columns = data.columns.str.lower()
```

Посчитаем количество пропусков

```python
#количество пропусков
display(pd.DataFrame(round(data.isna().mean()*100)).style.background_gradient('coolwarm'))
display(data.isna().sum())
```

**Есть пропуски и они могут повлиять на конечный результат, проводить исследование по таким данным нельзя, поэтому попытаемся обосновать наличие пропусков и по возможности их заполним**

### Проверка на дубликаты и добавление столбца общие продажи

Добавим столбец общие продажи

```python
#Общие продажи
data['total_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales'] 
#Либо data['total_sales'] = data[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis = 1)
```

```python
#проверка на явные дубликаты
data.duplicated().sum()
```

### Пропуски в году релиза (year_of_release)

Пропуски равномерны во всех платформах, и они появились в самых разнообразных играх по жанру и году выпуска, с 1990 по 2014 год. Некторые игры выпускались в один год на всех платформах одинаково, а другие были выпущены не зависимо от выхода на других платформах. Общая стоимость таких игр не повлияет на данные. Так что эти немногочисленные пропуски мы удалим.

```python
data.query('year_of_release.isna()==True')
```

```python
#Удалим пропуски с помощью среза
data = data.query('year_of_release.isna()==False')
```

### Пропуски в названии и жанре игры, (name, genge)

```python
data.query('genre.isna()==True')
```

Два пропуска в названии и жанре игры можно посмотреть и удалить, т.к. записи не помогут решить главную задачу проекта — на основе актуальных платформ сформировать рекомендации для проведения рекламной кампании в 2017 г. и само кол-во пропусков минимально

Удалим пропуски

```python
data = data.query('genre.isna()==False')
```

### Пропуски в  оценке критиков, оценки пользователей и рейтинге(critic_score, user_Score, rating)

Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта американская ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию c 1994 года. Пропуски в данных ни в коем случае нельзя удалять. 

```python
data.groupby('rating')['total_sales'].agg(['sum', 'count']).sort_values(by='count', ascending=False)
```

Видно, что данных в четырех нижних категориях мало, так что их лучше удалить, а вместо пропусков добавим категорию `unknown`, ведь они могут повлиять на данные

```python
data = data.query('rating != ["EC", "K-A", "AO", "RP"]')
data['rating'] = data['rating'].fillna('unknown')
```

```python
data.groupby('rating')['total_sales'].agg(['sum', 'count']).sort_values(by='count', ascending=False)
```

**В оценке пользователей (`user_score`) есть значение `tbd`(to be determind - будет определено), что говорит о том, что оценку еще не определили. Так что заменим значения `tbd` на пропуск, чтобы можно было поменять тип данных. `user_score` и `critic_score` данные object заменим их на float, чтобы с ними можно было работать. Переведем данные с типа float в тип int в `year_of_release`, так как год может быть только целочисленным.**

```python
for i in data['user_score']:
    data.loc[data['user_score'] == "tbd", 'user_score'] = np.nan
```

```python
#Перевод в тип int
data['year_of_release'] = data['year_of_release'].astype('int')
data['user_score'] = data['user_score'].astype('float')
data['critic_score'] = data['critic_score'].astype('float')
```

### **Промежуточный вывод:** 

*Данные подготовлены, пропуски удалить нельзя. Столбец `rating` избавили от пропусков. ***Явные дубликаты отсутствуют***.*

Таким образом можно предположить, что люди, cовершают насилие в виртуальном мире, что возможно успокаивает накопившийся стресс, ведь продажи довольно высоки. В общем и целом, лучшие категории для `NA` и `EU` -  `E`, `M`, а для Японии первое место занимают игры с добавленной категорией `unkonown`, делать выводу по рейтингу не стоит, ведь эти игры могут быть с любым рейтингом.

### **ПРОВЕРКА ГИПОТЕЗ**

#### Первая гипотеза


`H_0: Оценка (user_score) платформы "XOne" = оценка (user_score) платформы "PC"`

`H_a: Оценка (user_score) платформы "XOne" ≠ оценка (user_score) платформы "PC"
alpha = 0.05`

```python
XOne = good_data.query('platform=="XOne" and user_score.isna()==False')['user_score']
PC = good_data.query('platform=="PC" and user_score.isna()==False')['user_score']

alpha = 0.05
results = st.ttest_ind(XOne, PC, equal_var=False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**с вероятностью около 14,75% различие в средних оценках для исследуемых платформ мы получим случайно. Это явно слишком большая вероятность, чтобы делать вывод о значимом различии между средними оценками**

#### Вторая гипотеза

`H_0: Оценка (user_score) жанра "Action" = оценка (user_score) жанра "Sports"
alpha = 0.05`

`H_a: Оценка (user_score) жанра "Action" ≠ оценка (user_score) жанра "Sports"
alpha = 0.05`

```python
action = good_data.query('genre=="Action" and user_score.isna()==False')['user_score']
sports = good_data.query('genre=="Sports" and user_score.isna()==False')['user_score']

alpha = 0.05
results = st.ttest_ind(action, sports, equal_var=False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
def graph (df, year, region, name, axes):
    
    df = df.query('year_of_release >= @year')
    
    sales = df.pivot_table(index='platform', 
                           values=region, 
                           aggfunc='sum').nlargest(5, region)
    
    sales = sales.reset_index()
    
    sales = (
            sales.append({'platform': 'Other', region: df[region].sum() 
                       - sales[region].sum()}, ignore_index= True)
         )
    
    
    sales.columns = ['platform', 'sales']
      
    labels_c=sales.platform
    colours = {'Wii':'C0', 'NES':'C1', 'GB':'C2', 'DS':'C3', 'X360':'C4', 
    'PS3':'C5', 'PS2':'C6', 'SNES':'C7', 'GBA':'C8',
               'PS4':'steelblue', '3DS':'orange', 
               'N64':'C11', 'PS':'C12', 'XB':'C13', 'PC':'C14', '2600':'C15', 'PSP':'C16', 
               'XOne':'C17',
               'WiiU':'C18', 'GC':'C19', 'GEN':'C20', 'DC':'C21', 'PSV':'C22', 
               'SAT':'C23', 'SCD':'C24', 'WS':'C25', 'NG':'C26', 
               'TG16':'C27', '3DO':'C28', 'GG':'C29', 'PCFX':'C30', 'Other':'darkred'}
   
    sales.plot(kind='pie',
               y="sales",
               
               autopct='%1.0f%%',
               wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
               textprops={'size': 'x-large'}, 
               labels= labels_c,
               colors=[colours[key] for key in labels_c],
               legend=False, 
               title = f"Популярность платформ в {name} ", 
               ax = axes).set(ylabel='')
    
    plt.tight_layout()
```
